```python
import os
import pandas as pd
import seaborn as sns
from dotenv import load_dotenv
import spotipy
from spotipy.oauth2 import SpotifyOAuth

# load the .env file variables
load_dotenv()

client_id = os.environ.get("CLIENT_ID")
client_secret = os.environ.get("CLIENT_SECRET")

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id, client_secret=client_secret, redirect_uri="https://localhost"))


```

```python
id_artista= '05fG473iIaoy82BF1aGhL8'

respuesta = sp.artist_top_tracks(id_artista)

tracks = (respuesta['tracks'])
```

```python
df_tracks = pd.DataFrame.from_records(tracks)
df_tracks
```

```python
df_tracks.sort_values(['popularity'], inplace=True)
df_tracks
```

```python
df_tracks.head(3)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x=df_tracks['popularity'],y=df_tracks['duration_ms'])

plt.show()
```

```python
df = df_tracks[['name', 'popularity', 'duration_ms']]

# Convertir duracin de str a int
df['duration_ms'] = df['duration_ms'].astype(int)

# Convertir milisegundos a minutos y segundos
df['minutos'] = df['duration_ms'] // 60000  # 1 minuto = 60000 milisegundos
df['segundos'] = (df['duration_ms'] // 1000) % 60  # Segundos restantes despus de convertir a minutos

# Formatear la duracin como mm:ss
df['duration'] = df.apply(lambda row: '{:02d}:{:02d}'.format(row['minutos'], row['segundos']), axis=1)

# Eliminar las columnas temporales si lo deseas
df = df.drop(['minutos', 'segundos','duration_ms'], axis=1)

# Mostrar el DataFrame resultante
print(df)
```
